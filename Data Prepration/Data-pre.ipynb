```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import pandas as pd

# 1. Load FinBERT once
tokenizer = AutoTokenizer.from_pretrained("yiyanghkust/finbert-tone")
model     = AutoModelForSequenceClassification.from_pretrained("yiyanghkust/finbert-tone")
model.eval()

MAX_LEN = 512

def get_sentiment_probs(text: str):
    inputs = tokenizer(
        text,
        return_tensors="pt",
        truncation=True,
        max_length=MAX_LEN
    )
    with torch.no_grad():
        logits = model(**inputs).logits
        probs  = torch.softmax(logits, dim=1).squeeze().tolist()
    # probs = [P(negative), P(neutral), P(positive)]
    return probs

# 2. Apply to DataFrame and expand into three columns
prob_df = df['text'].apply(get_sentiment_probs).apply(
    lambda p: pd.Series(p, index=['p_neg', 'p_neu', 'p_pos'])
)
df = pd.concat([df, prob_df], axis=1)

# 3. Build alternative sentiment features:
#    A) Direct polarity = P_pos – P_neg
df['sent_polarity'] = df['p_pos'] - df['p_neg']

#    B) Thresholded flag for strong signals
threshold = 0.8
df['sent_strong'] = 0
df.loc[df['p_pos'] > threshold, 'sent_strong'] = +1
df.loc[df['p_neg'] > threshold, 'sent_strong'] = -1

#    C) Soft neutral vs. extreme weighting
df['sent_weighted'] = df['p_pos'] * 1 - df['p_neg'] * 1

# 4. Aggregate to daily level as you prefer:
daily = df.groupby('published_day').agg({
    'sent_polarity': 'mean',
    'sent_strong':   'mean',
    'sent_weighted':'mean'
})
```

```python
import yfinance as yf
import pandas as pd
import time

def fetch_price_data_for_dates(ticker, dates, retries=3, delay=5):
    """
    Download daily close prices for exactly the dates in `dates`.
    Retries up to `retries` times with `delay` seconds between attempts.
    Returns a DataFrame with columns ['date', '{ticker}_close'].
    """
    # Convert your date‐strings into pandas Timestamps
    ts_dates = [pd.to_datetime(d) for d in dates]
    start_ts = min(ts_dates)
    end_ts   = max(ts_dates)
    
    # Format for yfinance: include end_ts + 1 day so that end_ts itself is fetched
    start_str = start_ts.strftime("%Y-%m-%d")
    end_str   = (end_ts + pd.Timedelta(days=1)).strftime("%Y-%m-%d")

    for attempt in range(1, retries+1):
        try:
            data = yf.download(
                ticker,
                start=start_str,
                end=end_str,
                progress=False
            )
            if data.empty:
                raise ValueError("Downloaded DataFrame is empty")
            
            # Flatten MultiIndex columns if present
            if isinstance(data.columns, pd.MultiIndex):
                data.columns = data.columns.get_level_values(0)
            
            # Bring Date into a column and cast to string
            data = data.reset_index()
            data['date'] = data['Date'].dt.date.astype(str)
            
            # Keep only the rows you asked for
            data = data[data['date'].isin(dates)]
            
            # Return only date + Close, renamed to ticker_close
            return data[['date', 'Close']].rename(
                columns={'Close': f'{ticker}_close'}
            )
        
        except Exception as e:
            print(f"Attempt {attempt} failed for {ticker}: {e}")
            time.sleep(delay)
    
    raise RuntimeError(f"Failed to fetch data for {ticker} after {retries} attempts")

# — Step 1: Normalize your df’s published_day and get unique date-strings
df['published_day'] = pd.to_datetime(df['published_day'], errors='coerce').dt.date
unique_dates = df['published_day'].dropna().astype(str).unique().tolist()

# — Step 2: Fetch S&P500 and VIX close data for exactly those days
sp500_df = fetch_price_data_for_dates('^GSPC', unique_dates)
vix_df   = fetch_price_data_for_dates('^VIX', unique_dates)

# — Step 3: Merge the fetched close prices back into your original df
df['pub_day_str'] = df['published_day'].astype(str)

df = (
    df
    .merge(sp500_df.rename(columns={'date': 'pub_day_str'}), on='pub_day_str', how='left')
    .merge(vix_df  .rename(columns={'date': 'pub_day_str'}), on='pub_day_str', how='left')
)

# — Step 4: Clean up
df.drop(columns=['pub_day_str'], inplace=True)
```

```python
df = pd.read_csv('Financial_News_Articles.csv')
df['date'] = pd.to_datetime(df['published_day']).dt.normalize()
df = df.sort_values('date')
```

```python
# One row per calendar date
daily = (
    df[['date','^GSPC_close','^VIX_close']]
      .drop_duplicates()
      .set_index('date')
      .sort_index()
)


# **Now** assign the return columns
daily['sp500_ret'] = daily['^GSPC_close'].pct_change()
daily['vix_chg1']  = daily['^VIX_close'].pct_change()

# Forward‐fill weekends/holidays (optional)
daily = daily.asfreq('B').ffill()

# Check you have them
print("daily columns:", daily.columns.tolist())
# Expect: ['^GSPC_close','^VIX_close','sp500_ret','vix_chg1']
print(daily.head(), "\n")

df['date'] = pd.to_datetime(df['published_day'])
df = df.merge(
    daily[['sp500_ret','vix_chg1']],
    left_on='date',
    right_index=True,
    how='left'
)
```

```python
# ——————————————————————————————————————————————
# 4. Build lagged and rolling features
# ——————————————————————————————————————————————
# df['ret_lag1']   = df['sp500_ret'].shift(1)
# df['ret_roll5']  = df['sp500_ret'].rolling(window=5).mean().shift(1)
# df['vix_lag1']   = df['^VIX_close'].shift(1)
# df['vix_chg1']   = df['^VIX_close'].pct_change().shift(1)
# df['sent_lag1']  = pd.to_numeric(df['sentiment_numeric'], errors='coerce').shift(1)
# df['sent_roll3'] = pd.to_numeric(df['sentiment_numeric'], errors='coerce').rolling(window=3).mean().shift(1)


# 4. Build lagged and rolling features (including new sentiment cols)
df['ret_lag1']    = df['sp500_ret'].shift(1)
df['ret_roll5']   = df['sp500_ret'].rolling(5).mean().shift(1)
# df['vix_lag1']    = df['vix_close'].shift(1)                    # renamed for consistency?
# df['vix_chg1']    = df['vix_close'].pct_change().shift(1)

df['vix_lag1']  = df['^VIX_close'].shift(1)
df['vix_chg1']  = df['^VIX_close'].pct_change().shift(1)


# Replace sentiment_numeric with your new features:
df['polarity_lag1']   = df['sent_polarity'].shift(1)
df['polarity_roll3']  = df['sent_polarity'].rolling(3).mean().shift(1)

df['strong_lag1']     = df['sent_strong'].shift(1)
df['strong_roll3']    = df['sent_strong'].rolling(3).mean().shift(1)
df['polarity_x_strong'] = df['sent_polarity'] * df['sent_strong']
df['polarity_x_strong_lag1']     = df['polarity_x_strong'].shift(1)
df['polarity_x_strong_roll3']    = df['polarity_x_strong'].rolling(3).mean().shift(1)


# ——————————————————————————————————————————————
# 5. Drop incomplete rows (e.g., from shifts/rolls)
# ——————————————————————————————————————————————
feature_cols = ['ret_lag1','ret_roll5','vix_lag1','vix_chg1','polarity_lag1','polarity_roll3','strong_lag1','strong_roll3']


df = df.dropna(subset=['y'] + feature_cols)

# ——————————————————————————————————————————————
# 6. Standardize features for MCMC-friendly priors
# ——————————————————————————————————————————————
scaler = StandardScaler()
df[feature_cols] = scaler.fit_transform(df[feature_cols])
```
